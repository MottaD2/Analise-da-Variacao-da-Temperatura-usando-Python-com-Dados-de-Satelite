# Introdu√ß√£o e Fundamentos

#### **O que s√£o dados de sat√©lite e imagem de sat√©lite?**
Dados de sat√©lite, s√£o informa√ß√µes coletadas por sensores a bordo de sat√©lites que medem a temperatura da superf√≠cie terrestre, reflect√¢ncia em diferentes bandas, umidade do solo etc e as imagens s√£o as imagens da totalidade ou de parte da Terra, que s√£o obtidas atrav√©s dos sensores.

Um exemplo de instrumento √© o MODIS, o MODIS (ou Espectrorradi√¥metro de Imagem de Resolu√ß√£o Moderada) √© um instrumento essencial a bordo dos sat√©lites Terra (originalmente conhecido como EOS AM-1) e Aqua (originalmente conhecido como EOS PM-1). A √≥rbita do Terra ao redor da Terra √© cronometrada de modo que passe de norte a sul sobre o equador pela manh√£, enquanto o Aqua passa de sul a norte sobre o equador √† tarde. O Terra MODIS e o Aqua MODIS observam toda a superf√≠cie da Terra a cada 1 ou 2 dias, adquirindo dados em 36 bandas espectrais, ou grupos de comprimentos de onda (consulte as Especifica√ß√µes T√©cnicas do MODIS). Esses dados aprimorar√£o nossa compreens√£o da din√¢mica e dos processos globais que ocorrem na terra, nos oceanos e na baixa atmosfera.

Para a visualizar estes dados, geralmente os organizamos/obtemos o dado como um raster.


#### **O que √© um raster:**
Raster ou dados ‚Äúem grade‚Äù s√£o armazenados como uma grade de valores que s√£o representados em um mapa como pixels. Cada valor de pixel representa uma √°rea na superf√≠cie da Terra. Um arquivo raster √© composto por uma grade regular de c√©lulas, todas do mesmo tamanho.

Voc√™ j√° viu e usou rasters antes se j√° olhou para fotografias ou imagens em uma ferramenta como o Google Earth. No entanto, os arquivos raster com os quais voc√™ ir√° trabalhar s√£o diferentes das fotografias, pois s√£o espacialmente referenciados. Cada pixel representa uma √°rea de terreno no solo. Essa √°rea √© definida pela resolu√ß√£o espacial do raster.

#### **O que √© um vetor:**
Al√©m dos rasters, dentro da an√°lise e manipula√ß√£o de dados, tamb√©m lidamos com dados vetoriais.
Os dados vetoriais s√£o compostos de localiza√ß√µes geom√©tricas discretas (valores x, y) conhecidas como v√©rtices que definem a "forma" do objeto espacial. A organiza√ß√£o dos v√©rtices determina o tipo de vetor com o qual voc√™ est√° trabalhando. Existem tr√™s tipos de dados vetoriais:

**Pontos:** Cada ponto individual √© definido por uma √∫nica coordenada x, y. Pode haver muitos pontos em um arquivo de pontos vetoriais. Exemplos de dados pontuais incluem: locais de amostragem, localiza√ß√£o de √°rvores individuais ou localiza√ß√£o de parcelas.

**Linhas:** As linhas s√£o compostas de muitos (pelo menos 2) v√©rtices, ou pontos, que est√£o conectados. Por exemplo, uma estrada ou um riacho pode ser representado por uma linha. Esta linha √© composta por uma s√©rie de segmentos, cada "curva" na estrada ou c√≥rrego representa um v√©rtice que definiu x, y localiza√ß√£o.

**Pol√≠gonos:** Um pol√≠gono consiste em 3 ou mais v√©rtices conectados e "fechados". Assim, os contornos dos limites das parcelas, lagos, oceanos e estados ou pa√≠ses s√£o frequentemente representados por pol√≠gonos.

## **Matplotlib**
Cria√ß√£o de gr√°ficos, mapas e imagens, essencial para visualiza√ß√£o de dados de sat√©lite e vetoriais;

Matplot √© uma das principais bibliotecas utilizadas para plotar imagens, gr√°ficos e afins. Vamos plotar o nosso raster novamente:

E se quisermos plotar um shapefile junto com o nosso raster? Tamb√©m √© poss√≠vel! Mas antes algumas verifica√ß√µes precisam ser realizadas para que tudo funcione corretamente.

```python
# Verifique o CRS do shapefile e do raster

print(shapefile.crs)
print(modis_raster.rio.crs)
```

Como podemos observar, os nossos dados n√£o possuem o mesmo Sistema de Refer√™ncia de Coordenadas. Quando os dados n√£o est√£o no mesmo CRS, √© necess√°rio reprojet√°-los para que seja poss√≠vel plotar os dados juntos!

```python
# Armazene o crs do raster
crs = modis_raster.rio.crs

# Reprojete o shapefile
shapefile_reproj = shapefile.to_crs(crs)

# Verifique se o shapefile foi reprojetado
print(shapefile_reproj.crs)
```

Al√©m da reproje√ß√£o, √© importante utilizar a extens√£o do raster para que o plot funcione corretamente.

```python
# Obter extens√£o do raster (xmin, ymin, xmax, ymax)
xmin, ymin, xmax, ymax = modis_raster.rio.bounds()
extent = (xmin, xmax, ymin, ymax)
```

E agora vamos plotar.

```python
import matplotlib.pyplot as plt
import geopandas as gpd


# Plotar raster e shapefile
fig, ax = plt.subplots(figsize=(8,6))
plt.imshow(modis_raster.squeeze(), cmap="terrain", extent=extent)
shapefile_reproj.boundary.plot(ax=ax, color="red", linewidth=2)
plt.title("Raster MODIS + Shapefile Reprojetado")
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.colorbar(label="Valor do Pixel")
plt.show()
```

O nosso shapefile √© muito maior que o nosso raster, ent√£o quando fazemos o plot o raster quase n√£o aparece. A solu√ß√£o para este problema √© recortar o nosso shapefile pela √°rea do raster

```python
import geopandas as gpd
from shapely.geometry import box

# --- Criar pol√≠gono da √°rea de cobertura do raster ---
xmin, ymin, xmax, ymax = modis_raster.rio.bounds()

raster_box = box(xmin, ymin, xmax, ymax)

raster_gdf = gpd.GeoDataFrame({"geometry": [raster_box]}, crs=modis_raster.rio.crs)

# --- Recortar shapefile pela √°rea do raster ---
shapefile_clip = gpd.overlay(shapefile_reproj, raster_gdf, how='intersection')


```

Vamos plotar o shapefile recortado

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize=(8,6))

shapefile_clip.plot(ax =ax, color ="none", edgecolor = "green", linewidth = 2)
plt.title("Shapefile recortado pela √°rea do Raster")
plt.show()
```

E agora sim realizamos novamente o plot do shapefile com o raster:

```python
import matplotlib.pyplot as plt

extent = (xmin, xmax, ymin, ymax)

fig, ax = plt.subplots(figsize=(8,6))

plt.imshow(modis_raster.squeeze(), cmap="terrain", extent=extent)

shapefile_clip.boundary.plot(ax=ax, color='red', linewidth=2)

plt.title("Shapefile recortado pela √°rea do Raster")
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.colorbar(label="Valor do Pixel")
plt.show()
```

## Introdu√ß√£o √†s cole√ß√µes
No Cat√°logo de Dados do Earth Engine, os conjuntos de dados podem ser de diferentes tipos:

* *Caracter√≠sticas* que s√£o objetos geom√©tricos com uma lista de propriedades. Por exemplo, uma bacia hidrogr√°fica com algumas propriedades, como nome e √°rea , √© um ee.Feature.
* *Imagens* que s√£o como caracter√≠sticas, mas podem incluir v√°rias faixas. Por exemplo, a eleva√ß√£o do solo fornecida pelo USGS aqui √© um ee.Image.
* *Cole√ß√µes* que s√£o grupos de fei√ß√µes ou imagens. Por exemplo, as Camadas de Unidades Administrativas Globais que fornecem limites administrativos s√£o um ee.FeatureCollectione o conjunto de dados de Temperatura da Superf√≠cie Terrestre MODIS √© um ee.ImageCollection.
Se quiser saber mais sobre diferentes modelos de dados, visite o Guia do Usu√°rio do Earth Engine .

Nas se√ß√µes a seguir, trabalhamos com a cobertura do solo (LC) do MODIS , a temperatura da superf√≠cie do solo (LST) do MODIS e com a eleva√ß√£o do solo (ELV) do USGS , que s√£o ee.ImageCollections. As descri√ß√µes dos conjuntos de dados nos fornecem todas as informa√ß√µes necess√°rias para importar e manipular esses conjuntos de dados: a disponibilidade, o provedor, o snippet do Earth Engine e as bandas dispon√≠veis associadas √†s imagens na cole√ß√£o.

Agora, para importar as cole√ß√µes LC, LST e ELV, podemos copiar e colar os trechos do Earth Engine:

```python
# Importar cole√ß√µes MODIS
lc = ee.ImageCollection('MODIS/061/MCD12Q1')   # Land Cover Type (vers√£o 061)
lst = ee.ImageCollection('MODIS/061/MOD11A1')  # Land Surface Temperature (vers√£o 061)

# Eleva√ß√£o
elv = ee.Image('USGS/SRTMGL1_003')
```

Todas essas imagens v√™m em diferentes resolu√ß√µes, frequ√™ncias e, possivelmente, proje√ß√µes, variando de imagens di√°rias com resolu√ß√£o de 1 km para o LST (da√≠ o nome "an" ee.ImageCollection‚Äî uma cole√ß√£o de v√°rias imagens ee.Images) a uma √∫nica imagem representando dados do ano 2000 com resolu√ß√£o de 30 m para o ELV. Embora precisemos ficar de olho na frequ√™ncia, o GEE cuida da resolu√ß√£o e da proje√ß√£o, reamostrando e reprojetando todos os dados com os quais trabalharemos em uma proje√ß√£o comum (saiba mais sobre proje√ß√µes no Earth Engine ). Podemos definir a resolu√ß√£o (chamada de escala no GEE) sempre que necess√°rio e, claro, temos a op√ß√£o de n√£o for√ßar nenhuma reproje√ß√£o.

Como voc√™ pode ver na descri√ß√£o dos conjuntos de dados, eles incluem diversos conjuntos de informa√ß√µes armazenados em diversas faixas. Por exemplo, essas faixas est√£o associadas √† cole√ß√£o LST:

* *LST_Day_1km* : Temperatura da superf√≠cie terrestre durante o dia
* *Day_view_time* : hora local de observa√ß√£o do dia
* *LST_Night_1km* : Temperatura da superf√≠cie terrestre noturna
etc.
A p√°gina de descri√ß√£o da cole√ß√£o nos informa que o nome da banda associada ao LST diurno √© LST_Day_1km , em unidades Kelvin. Al√©m disso, os valores variam de 7.500 a 65.535, com uma escala corretiva de 0,02.

Em seguida, precisamos filtrar a coleta pelo per√≠odo desejado. Podemos fazer isso usando o m√©todo filterDate() . Tamb√©m precisamos selecionar as faixas com as quais queremos trabalhar. Portanto, decidimos focar no LST diurno, ent√£o selecionamos a faixa diurna LST_Day_1km e seu indicador de qualidade associado QC_Day com o m√©todo select().

```python
# Data inicial de interesse (inclusive).
i_date = '2017-01-01'

# Data final de interesse (exclusivo).
f_date = '2020-01-01'

# Sele√ß√£o de bandas e datas apropriadas para o LST.
lst = lst.select('LST_Day_1km', 'QC_Day').filterDate(i_date, f_date)
```

Agora, podemos carregar arquivos de forma existentes ou definir alguns pontos com coordenadas de longitude e latitude onde queremos saber mais sobre LC, LST e ELV. Para este exemplo, vamos usar duas localiza√ß√µes de pontos:

Podemos obter facilmente informa√ß√µes sobre nossa regi√£o/ponto de interesse usando os seguintes m√©todos (para obter mais informa√ß√µes sobre os m√©todos dispon√≠veis e os argumentos necess√°rios, visite a documenta√ß√£o da API aqui ):

* sample(): amostra a imagem (N√ÉO funciona para um ee.ImageCollection‚Äî falaremos sobre amostragem do ee.ImageCollection) de acordo com uma geometria dada e uma escala (em metros) da proje√ß√£o a ser amostrada. Ele retorna um ee.FeatureCollection.

* get(): para selecionar a banda apropriada da sua Imagem/Cole√ß√£o,
* getInfo(): avalia o gr√°fico de express√£o do lado do servidor e transfere o resultado para o cliente.
Em seguida, podemos consultar a eleva√ß√£o do solo e o LST em torno do nosso ponto de interesse usando os seguintes comandos. Tenha cuidado ao avaliar o LST. De acordo com a descri√ß√£o do conjunto de dados , o valor deve ser corrigido por um fator de 0,02 para obter unidades Kelvin (n√£o se esque√ßa da convers√£o). Para obter o LST m√©dio diurno plurianual, usamos o m√©todo mean() de redu√ß√£o de coleta no LST ee.ImageCollection.

```python
scale = 1000  # escala em metros

# Imprima a eleva√ß√£o perto de Lyon, Fran√ßa.
elv_urban_point = elv.sample(u_poi, scale).first().get('elevation').getInfo()
print('Eleva√ß√£o do terreno no ponto urbano:', elv_urban_point, 'm')

# Calcule e imprima o valor m√©dio da cole√ß√£o de LST no ponto.
lst_urban_point = lst.mean().sample(u_poi, scale).first().get('LST_Day_1km').getInfo()
print('LST m√©dia durante o dia no ponto urbano:', round(lst_urban_point*0.02 -273.15, 2), '¬∞C')

# Imprima o tipo de cobertura do solo no ponto.
lc_urban_point = lc.first().sample(u_poi, scale).first().get('LC_Type1').getInfo()
print('O valor da cobertura do solo no ponto urbano √©:', lc_urban_point)
```

## Obter uma s√©rie temporal
Agora que voc√™ v√™ que podemos obter informa√ß√µes geoespaciais sobre um local de interesse com bastante facilidade, talvez voc√™ queira inspecionar uma s√©rie temporal, provavelmente criar alguns gr√°ficos e calcular estat√≠sticas sobre um local. Portanto, importamos os dados nos locais fornecidos usando o m√©todo getRegion()

```python
# Obtenha os dados do pixel que intercepta o ponto na √°rea urbana.
lst_u_poi = lst.getRegion(u_poi, scale).getInfo()

# Obtenha os dados do pixel que intercepta o ponto na √°rea rural.
lst_r_poi = lst.getRegion(r_poi, scale).getInfo()

# Visualize o resultado.
lst_u_poi[:5]
```

A impress√£o das primeiras 5 linhas do resultado mostra que agora temos matrizes repletas de dados. Como podemos ver, v√°rios Nonevalores aparecem na coluna LST_Day_1km . O indicador de qualidade associado, QC_Day, indica o valor 2 , o que significa que o LST n√£o √© calculado devido aos efeitos das nuvens.

Agora definimos uma fun√ß√£o para transformar esse array em um Dataframe pandas, que √© muito mais conveniente de manipular.

```python
import pandas as pd

def ee_array_to_df(arr, list_of_bands):
    """Transforma o array retornado pelo `ee.Image.getRegion` no lado do cliente em um `pandas.DataFrame`."""
    df = pd.DataFrame(arr)

    # Reorganize o cabe√ßalho.
    headers = df.iloc[0]
    df = pd.DataFrame(df.values[1:], columns=headers)

    # Remova as linhas que n√£o cont√™m dados.
    df = df[['longitude', 'latitude', 'time', *list_of_bands]].dropna()

    # Converta os dados para valores num√©ricos.
    for band in list_of_bands:
        df[band] = pd.to_numeric(df[band], errors='coerce')

    # Converta o campo de tempo em um formato datetime.
    df['datetime'] = pd.to_datetime(df['time'], unit='ms')

    # Mantenha apenas as colunas de interesse.
    df = df[['time','datetime',  *list_of_bands]]

    return df
```

Aplicamos esta fun√ß√£o para obter as duas s√©ries temporais que queremos (e imprimir uma).



```python
lst_df_urban = ee_array_to_df(lst_u_poi,['LST_Day_1km'])

def t_modis_to_celsius(t_modis):
    """Converte as unidades da LST do MODIS para graus Celsius."""
    t_celsius =  0.02*t_modis - 273.15
    return t_celsius

# Aplique a fun√ß√£o para obter a temperatura em Celsius.
lst_df_urban['LST_Day_1km'] = lst_df_urban['LST_Day_1km'].apply(t_modis_to_celsius)

# Fa√ßa o mesmo para o ponto rural.
lst_df_rural = ee_array_to_df(lst_r_poi,['LST_Day_1km'])
lst_df_rural['LST_Day_1km'] = lst_df_rural['LST_Day_1km'].apply(t_modis_to_celsius)

lst_df_urban.head()
```

Agora que temos nossos dados em bom estado, podemos facilmente criar gr√°ficos e comparar as tend√™ncias. Como a regi√£o de Lyon, na Fran√ßa, apresenta um clima semicontinental, esperamos que o LST tenha influ√™ncia sazonal e a tend√™ncia sinusoidal descrita por Stallman (1965) , conforme segue:

Captura de tela 2025-10-15 095842.png

Esse c√≥digo realiza um ajuste de curva senoidal aos dados de temperatura da superf√≠cie (LST) para √°reas urbanas e rurais, seguindo estes passos:

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy import optimize
%matplotlib inline

# Ajuste de curvas
## Primeiro, extraia os valores de x (tempos) dos dataframes.
x_data_u = np.asanyarray(lst_df_urban['time'].apply(float))  # urban
x_data_r = np.asanyarray(lst_df_rural['time'].apply(float))  # rural

## Em seguida, extraia os valores de y (LST) dos dataframes.
y_data_u = np.asanyarray(lst_df_urban['LST_Day_1km'].apply(float))  # urban
y_data_r = np.asanyarray(lst_df_rural['LST_Day_1km'].apply(float))  # rural

## Depois, defina a fun√ß√£o de ajuste com os par√¢metros.
def fit_func(t, lst0, delta_lst, tau, phi):
    return lst0 + (delta_lst/2)*np.sin(2*np.pi*t/tau + phi)

## Otimize os par√¢metros utilizando um bom palpite inicial `p0`.

lst0 = 20
delta_lst = 40
tau = 365*24*3600*1000   # Milissegundos em um ano
phi = 2*np.pi*4*30.5*3600*1000/tau  # Deslocamento em rela√ß√£o ao momento em que esperamos que LST(t) = LST‚ÇÄ


params_u, params_covariance_u = optimize.curve_fit(
    fit_func, x_data_u, y_data_u, p0=[lst0, delta_lst, tau, phi])
params_r, params_covariance_r = optimize.curve_fit(
    fit_func, x_data_r, y_data_r, p0=[lst0, delta_lst, tau, phi])
```

E agora plote!

```python

# Subplots.
fig, ax = plt.subplots(figsize=(14, 6))

# Adicione gr√°ficos de dispers√£o.
ax.scatter(lst_df_urban['datetime'], lst_df_urban['LST_Day_1km'],
           c='black', alpha=0.2, label='Urban (data)')
ax.scatter(lst_df_rural['datetime'], lst_df_rural['LST_Day_1km'],
           c='green', alpha=0.35, label='Rural (data)')

# Adicione as curvas de ajuste.

ax.plot(lst_df_urban['datetime'],
        fit_func(x_data_u, params_u[0], params_u[1], params_u[2], params_u[3]),
        label='Urban (fitted)', color='black', lw=2.5)
ax.plot(lst_df_rural['datetime'],
        fit_func(x_data_r, params_r[0], params_r[1], params_r[2], params_r[3]),
        label='Rural (fitted)', color='green', lw=2.5)

# Adicione alguns par√¢metros.
ax.set_title('Daytime Land Surface Temperature Near Lyon', fontsize=16)
ax.set_xlabel('Date', fontsize=14)
ax.set_ylabel('Temperature [C]', fontsize=14)
ax.set_ylim(-0, 40)
ax.grid(lw=0.2)
ax.legend(fontsize=14, loc='lower right')

plt.show()
```

## Outras formas de baixar imagens do GEE
O Google Earth Engine (GEE) n√£o se limita apenas √† obten√ß√£o de produtos derivados de s√©ries temporais, como temperaturas superficiais ou √≠ndices de vegeta√ß√£o. Ele oferece acesso a uma grande variedade de produtos provenientes de diversos sensores de sat√©lite, que permitem an√°lises espaciais e visuais detalhadas do ambiente terrestre.

Entre as possibilidades, destacam-se:

1 - Imagens de reflect√¢ncia:

* S√£o imagens que representam a quantidade de radia√ß√£o refletida pela superf√≠cie terrestre em diferentes comprimentos de onda.

* Com elas, √© poss√≠vel estudar a cobertura do solo, sa√∫de da vegeta√ß√£o, uso do solo, presen√ßa de √°gua, entre outros.

2 - Visualiza√ß√£o em RGB:

* O GEE permite combinar bandas espectrais para criar imagens coloridas em RGB verdadeiro (Red-Green-Blue) ou falso, facilitando a interpreta√ß√£o visual de √°reas de interesse.

* Por exemplo, em regi√µes agr√≠colas, florestais ou urbanas, √© poss√≠vel identificar diferentes tipos de cobertura do solo apenas observando a combina√ß√£o das cores.

```python
import ee
import requests
import rasterio
from io import BytesIO
import matplotlib.pyplot as plt
import numpy as np
```

```python

# Sua geometria e sele√ß√£o de bandas, como no seu c√≥digo:
aoi = ee.Geometry.Point(-57.90171, -6.16505)
roi = aoi.buffer(1e6)

bandas = ['sur_refl_b01', 'sur_refl_b07', 'sur_refl_b03']


region = [[-60.78199491764261,-3.8146864244855014],[-59.06812773014261,-3.8146864244855014],
[-59.06812773014261,-2.4818924114962577],[-60.78199491764261,-2.4818924114962577]]
scale = 1000

colecao = ee.ImageCollection("MODIS/061/MOD13A3").filterBounds(roi).filterDate('2001-08-15', '2001-10-30')
imagem = colecao.select(bandas).median()



```

E voc√™ pode carregar a imagem diretamente para a mem√≥ria. Observa√ß√£o: Para este m√©todo, o tamanho m√°ximo da solicita√ß√£o √© de 32 MB, e a dimens√£o m√°xima da grade √© de 10.000.

```python
# Gerar URL para download direto
url = imagem.getDownloadURL({
    'scale': scale,
    'region': region,
    'format': 'GEO_TIFF'
})
```

```python
# Baixar a imagem e carregar com rasterio em mem√≥ria
response = requests.get(url)
with rasterio.MemoryFile(response.content) as memfile:
    with memfile.open() as dataset:
        array = dataset.read()  # l√™ todas as bandas como numpy array
        profile = dataset.profile
```

```python
# Vamos garantir que os valores estejam normalizados para visualiza√ß√£o (entre 0 e 1)
def normalize(band):
    band = band.astype(np.float32)
    return (band - band.min()) / (band.max() - band.min())

# Seleciona 3 bandas para o RGB
# Exemplo: banda 1 = Vermelho, banda 2 = Infravermelho Pr√≥ximo (NIR), banda 3 = Azul

# Ajuste se necess√°rio conforme a ordem das bandas no array
r = normalize(array[0])
g = normalize(array[1])
b = normalize(array[2])

# Empilha para formar uma imagem RGB
rgb = np.stack([r, g, b], axis=-1)

# Plotar
plt.figure(figsize=(10, 10))
plt.imshow(rgb)
plt.axis('off')
plt.title('Composi√ß√£o RGB da imagem MODIS')
plt.show()
```

### Obter a imagem do **MODIS**

```python
import ee
import datetime
import matplotlib.pyplot as plt


# Coordenadas e produto [Defina a sua coordenada]
# Lembre-se de trocar a v√≠rgula por ponto!!!

lat = -21.4887618
lon = -41.3756182

# Produto MODIS
prod = 'MODIS/061/MOD11A2'
data_band = 'LST_Day_1km'
qc_band = 'QC_Day'

# Criar ponto e buffer (10 km ao redor)
point = ee.Geometry.Point([lon, lat])
roi = point.buffer(10000)  # 10 km em metros
```

Liste os anos para a observa√ß√£o

```python
 # 10 anos, ajuste conforme necess√°rio
 # Selecione os anos de interesse

anos = list(range(2010, 2020))
```

### Atividade 1 - Calcular e plotar m√©dia anual

**Objetivo**

Conseguir identificar atrav√©s dos dados e gr√°ficos, os anos mais frios e mais quentes dentro do anos definidos!

```python
# ========================================
# Atividade 1: Calcular e plotar m√©dia anual
# ========================================

# Fun√ß√£o para calcular m√©dia anual de LST
# [APENAS RODE!]
def mean_annual_lst(year):
    start = ee.Date.fromYMD(year, 1, 1)
    end = ee.Date.fromYMD(year, 12, 31)

    collection = ee.ImageCollection(prod) \
        .filterBounds(roi) \
        .filterDate(start, end) \
        .select(data_band)

    # Convertendo de LST MODIS para ¬∞C
    annual_mean = collection.mean().multiply(0.02).subtract(273.15)
    return annual_mean.set('year', year)
```

```python
# Calcula imagens m√©dias anuais
# [APENAS RODE!]
annual_images = [mean_annual_lst(y) for y in anos]
```

```python
# Fun√ß√£o para calcular m√©dia da ROI
# [APENAS RODE!]

def mean_temp(image):
    mean_dict = image.reduceRegion(
        reducer=ee.Reducer.mean(),
        geometry=roi,
        scale=1000,
        maxPixels=1e9
    )
    return ee.Feature(None, {'year': image.get('year'), 'mean_temp': mean_dict.get(data_band)})
```

```python
# Calcula m√©dias da ROI para cada ano
# [APENAS RODE!]

features = [mean_temp(img) for img in annual_images]
fc = ee.FeatureCollection(features)
```

Para cada imagem anual (annual_images), aplica a fun√ß√£o mean_temp Cria uma FeatureCollection, que √© como uma tabela no Earth Engine: cada linha = ano, cada coluna = m√©dia de temperatura.

Agora vamos extrair listas de anos e m√©dias


```python
# Extrair listas
# [APENAS RODE!]

temp_list = fc.aggregate_array('mean_temp').getInfo() # aggregate_array('mean_temp') ‚Üí cria uma lista Python com as m√©dias anuais.
year_list = fc.aggregate_array('year').getInfo() # aggregate_array('year') ‚Üí cria uma lista com os anos correspondentes.
                                                  # getInfo() ‚Üí converte do formato Earth Engine para uma lista Python.
```

Identificando anos mais frio e mais quente

```python
# Identificar anos mais frio e mais quente
#min(temp_list) ‚Üí encontra a menor m√©dia anual (mais frio).
#max(temp_list) ‚Üí encontra a maior m√©dia anual (mais quente).
#index() ‚Üí pega a posi√ß√£o do ano correspondente na lista de anos.


# [APENAS RODE!]

coldest_year = year_list[temp_list.index(min(temp_list))]
warmest_year = year_list[temp_list.index(max(temp_list))]

print(f"Ano mais frio: {coldest_year} ({min(temp_list):.2f} ¬∞C)")
print(f"Ano mais quente: {warmest_year} ({max(temp_list):.2f} ¬∞C)")
```

 Plotando o gr√°fico

```python
# Plotando gr√°fico de linha
plt.figure(figsize=(10,6))
plt.plot(year_list, temp_list, marker='o', linestyle='-', color='orange') # Configure as cores como desejar
plt.title('M√©dia Anual de Temperatura (¬∞C) ‚Äì ROI 10 km')
plt.xlabel('Ano')
plt.ylabel('Temperatura (¬∞C)')
plt.grid(True)

# Destacar ano mais frio e mais quente
coldest_year = year_list[temp_list.index(min(temp_list))]
warmest_year = year_list[temp_list.index(max(temp_list))]


plt.scatter(coldest_year, min(temp_list), color='blue', s=100, label=f'Mais frio: {coldest_year}') # Configure as cores como desejar
plt.scatter(warmest_year, max(temp_list), color='red', s=100, label=f'Mais quente: {warmest_year}') # Configure as cores como desejar
plt.legend()
plt.show()
```

### Atividade 2 ‚Äì An√°lise Sazonal (m√©dia mensal).
**Objetivo**

Analisar como a temperatura varia ao longo dos meses de um ano, ou seja, capturar a sazonalidade.

```python
# ========================================
# Atividade 2: An√°lise Sazonal (m√©dia mensal)
# ========================================
# Exemplo para 1 ano
# Define o ano que voc√™ quer analisar.
# Cria a lista de meses de 1 a 12.
# Inicializa uma lista vazia monthly_means para armazenar a m√©dia de cada m√™s.

year = 2017 # Escolha o ano
months = list(range(1,13))
monthly_means = []
```

Para cada m√™s:


```python
# [APENAS RODE]
for m in months:
    start = ee.Date.fromYMD(year, m, 1) #  start ‚Üí primeiro dia do m√™s
    end = start.advance(1, 'month') # end ‚Üí primeiro dia do m√™s seguinte

    # Isso cria um intervalo mensal para filtrar os dados MODIS. Filtra a cole√ß√£o MODIS LST para a ROI e para o m√™s espec√≠fico.
    collection = ee.ImageCollection(prod).filterBounds(roi).filterDate(start,end).select(data_band)

    # Calcula a m√©dia das imagens do m√™s.
    monthly_mean = collection.mean().multiply(0.02).subtract(273.15)

    # Multiplica por 0.02 e subtrai 273.15 para converter de Kelvin para ¬∞C.
    # Calcula a m√©dia espacial da ROI para o m√™s usando reduceRegion.
    mean_dict = monthly_mean.reduceRegion(reducer=ee.Reducer.mean(), geometry=roi, scale=1000, maxPixels=1e9)

    # Adiciona o valor √† lista monthly_means
    monthly_means.append(mean_dict.get(data_band).getInfo())
```

Plote um gr√°fico de linha da temperatura m√©dia mensal:





```python
# Plota um gr√°fico de linha da temperatura m√©dia mensal:

plt.figure(figsize=(10,6))
plt.plot(months, monthly_means, marker='o', linestyle='-', color='green') # Configure as cores como desejar
plt.title(f'M√©dia Mensal de Temperatura - {year}')
plt.xlabel('M√™s') # Eixo X ‚Üí meses (1 a 12)
plt.ylabel('Temperatura (¬∞C)') # Eixo Y ‚Üí temperatura m√©dia (¬∞C) Visualiza picos e vales de temperatura ao longo do ano.
plt.grid(True)
plt.show()
```

### Atividade 3 -  Tend√™ncia Temporal

**Objetivo**

Mostrar se a temperatura m√©dia anual est√° aumentando, diminuindo ou est√° est√°vel ao longo dos anos, ou seja, calcular uma linha de tend√™ncia (regress√£o linear) sobre os dados anuais.

```python
# ========================================
# Atividade 3: Tend√™ncia Temporal
# ========================================

# [APENAS RODE!]
import numpy as np
from numpy.polynomial.polynomial import polyfit

# Ajuste linear
coef = polyfit(year_list, temp_list, 1)  # coef[0]=intercepto, coef[1]=inclina√ß√£o
trend = [coef[1]*y + coef[0] for y in year_list]

```

```python
# Plote a m√©dia anual com o dado year_list

plt.figure(figsize=(10,6))

plt.plot(year_list,
         temp_list,
         marker='o',
         linestyle='-',
         color='orange', # Configure as cores como desejar
         label='M√©dia anual')
plt.plot(year_list,
         trend,
         color='red', # Configure as cores como desejar
         linestyle='--',
         label=f'Tend√™ncia ({coef[1]:.3f} ¬∞C/ano)')
plt.title('Tend√™ncia de Temperatura Anual')
plt.xlabel('Ano')
plt.ylabel('Temperatura (¬∞C)')
plt.grid(True)
plt.legend()
plt.show()
```
