# Introdução e Fundamentos

#### **O que são dados de satélite e imagem de satélite?**
Dados de satélite, são informações coletadas por sensores a bordo de satélites que medem a temperatura da superfície terrestre, reflectância em diferentes bandas, umidade do solo etc e as imagens são as imagens da totalidade ou de parte da Terra, que são obtidas através dos sensores.

Um exemplo de instrumento é o MODIS, o MODIS (ou Espectrorradiômetro de Imagem de Resolução Moderada) é um instrumento essencial a bordo dos satélites Terra (originalmente conhecido como EOS AM-1) e Aqua (originalmente conhecido como EOS PM-1). A órbita do Terra ao redor da Terra é cronometrada de modo que passe de norte a sul sobre o equador pela manhã, enquanto o Aqua passa de sul a norte sobre o equador à tarde. O Terra MODIS e o Aqua MODIS observam toda a superfície da Terra a cada 1 ou 2 dias, adquirindo dados em 36 bandas espectrais, ou grupos de comprimentos de onda (consulte as Especificações Técnicas do MODIS). Esses dados aprimorarão nossa compreensão da dinâmica e dos processos globais que ocorrem na terra, nos oceanos e na baixa atmosfera.

Para a visualizar estes dados, geralmente os organizamos/obtemos o dado como um raster.


#### **O que é um raster:**
Raster ou dados “em grade” são armazenados como uma grade de valores que são representados em um mapa como pixels. Cada valor de pixel representa uma área na superfície da Terra. Um arquivo raster é composto por uma grade regular de células, todas do mesmo tamanho.

Você já viu e usou rasters antes se já olhou para fotografias ou imagens em uma ferramenta como o Google Earth. No entanto, os arquivos raster com os quais você irá trabalhar são diferentes das fotografias, pois são espacialmente referenciados. Cada pixel representa uma área de terreno no solo. Essa área é definida pela resolução espacial do raster.

#### **O que é um vetor:**
Além dos rasters, dentro da análise e manipulação de dados, também lidamos com dados vetoriais.
Os dados vetoriais são compostos de localizações geométricas discretas (valores x, y) conhecidas como vértices que definem a "forma" do objeto espacial. A organização dos vértices determina o tipo de vetor com o qual você está trabalhando. Existem três tipos de dados vetoriais:

**Pontos:** Cada ponto individual é definido por uma única coordenada x, y. Pode haver muitos pontos em um arquivo de pontos vetoriais. Exemplos de dados pontuais incluem: locais de amostragem, localização de árvores individuais ou localização de parcelas.

**Linhas:** As linhas são compostas de muitos (pelo menos 2) vértices, ou pontos, que estão conectados. Por exemplo, uma estrada ou um riacho pode ser representado por uma linha. Esta linha é composta por uma série de segmentos, cada "curva" na estrada ou córrego representa um vértice que definiu x, y localização.

**Polígonos:** Um polígono consiste em 3 ou mais vértices conectados e "fechados". Assim, os contornos dos limites das parcelas, lagos, oceanos e estados ou países são frequentemente representados por polígonos.

E se quisermos plotar um shapefile junto com o nosso raster? Também é possível! Mas antes algumas verificações precisam ser realizadas para que tudo funcione corretamente.

```python
# Verifique o CRS do shapefile e do raster

print(shapefile.crs)
print(modis_raster.rio.crs)
```

Como podemos observar, os nossos dados não possuem o mesmo Sistema de Referência de Coordenadas. Quando os dados não estão no mesmo CRS, é necessário reprojetá-los para que seja possível plotar os dados juntos!

```python
# Armazene o crs do raster
crs = modis_raster.rio.crs

# Reprojete o shapefile
shapefile_reproj = shapefile.to_crs(crs)

# Verifique se o shapefile foi reprojetado
print(shapefile_reproj.crs)
```

Além da reprojeção, é importante utilizar a extensão do raster para que o plot funcione corretamente.

```python
# Obter extensão do raster (xmin, ymin, xmax, ymax)
xmin, ymin, xmax, ymax = modis_raster.rio.bounds()
extent = (xmin, xmax, ymin, ymax)
```

E agora vamos plotar.

```python
import matplotlib.pyplot as plt
import geopandas as gpd


# Plotar raster e shapefile
fig, ax = plt.subplots(figsize=(8,6))
plt.imshow(modis_raster.squeeze(), cmap="terrain", extent=extent)
shapefile_reproj.boundary.plot(ax=ax, color="red", linewidth=2)
plt.title("Raster MODIS + Shapefile Reprojetado")
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.colorbar(label="Valor do Pixel")
plt.show()
```

O nosso shapefile é muito maior que o nosso raster, então quando fazemos o plot o raster quase não aparece. A solução para este problema é recortar o nosso shapefile pela área do raster

```python
import geopandas as gpd
from shapely.geometry import box

# --- Criar polígono da área de cobertura do raster ---
xmin, ymin, xmax, ymax = modis_raster.rio.bounds()

raster_box = box(xmin, ymin, xmax, ymax)

raster_gdf = gpd.GeoDataFrame({"geometry": [raster_box]}, crs=modis_raster.rio.crs)

# --- Recortar shapefile pela área do raster ---
shapefile_clip = gpd.overlay(shapefile_reproj, raster_gdf, how='intersection')


```

Vamos plotar o shapefile recortado

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize=(8,6))

shapefile_clip.plot(ax =ax, color ="none", edgecolor = "green", linewidth = 2)
plt.title("Shapefile recortado pela área do Raster")
plt.show()
```

E agora sim realizamos novamente o plot do shapefile com o raster:

```python
import matplotlib.pyplot as plt

extent = (xmin, xmax, ymin, ymax)

fig, ax = plt.subplots(figsize=(8,6))

plt.imshow(modis_raster.squeeze(), cmap="terrain", extent=extent)

shapefile_clip.boundary.plot(ax=ax, color='red', linewidth=2)

plt.title("Shapefile recortado pela área do Raster")
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.colorbar(label="Valor do Pixel")
plt.show()
```

## Introdução às coleções
No Catálogo de Dados do Earth Engine, os conjuntos de dados podem ser de diferentes tipos:

* *Características* que são objetos geométricos com uma lista de propriedades. Por exemplo, uma bacia hidrográfica com algumas propriedades, como nome e área , é um ee.Feature.
* *Imagens* que são como características, mas podem incluir várias faixas. Por exemplo, a elevação do solo fornecida pelo USGS aqui é um ee.Image.
* *Coleções* que são grupos de feições ou imagens. Por exemplo, as Camadas de Unidades Administrativas Globais que fornecem limites administrativos são um ee.FeatureCollectione o conjunto de dados de Temperatura da Superfície Terrestre MODIS é um ee.ImageCollection.
Se quiser saber mais sobre diferentes modelos de dados, visite o Guia do Usuário do Earth Engine .

Nas seções a seguir, trabalhamos com a cobertura do solo (LC) do MODIS , a temperatura da superfície do solo (LST) do MODIS e com a elevação do solo (ELV) do USGS , que são ee.ImageCollections. As descrições dos conjuntos de dados nos fornecem todas as informações necessárias para importar e manipular esses conjuntos de dados: a disponibilidade, o provedor, o snippet do Earth Engine e as bandas disponíveis associadas às imagens na coleção.

Agora, para importar as coleções LC, LST e ELV, podemos copiar e colar os trechos do Earth Engine:

```python
# Importar coleções MODIS
lc = ee.ImageCollection('MODIS/061/MCD12Q1')   # Land Cover Type (versão 061)
lst = ee.ImageCollection('MODIS/061/MOD11A1')  # Land Surface Temperature (versão 061)

# Elevação
elv = ee.Image('USGS/SRTMGL1_003')
```

Todas essas imagens vêm em diferentes resoluções, frequências e, possivelmente, projeções, variando de imagens diárias com resolução de 1 km para o LST (daí o nome "an" ee.ImageCollection— uma coleção de várias imagens ee.Images) a uma única imagem representando dados do ano 2000 com resolução de 30 m para o ELV. Embora precisemos ficar de olho na frequência, o GEE cuida da resolução e da projeção, reamostrando e reprojetando todos os dados com os quais trabalharemos em uma projeção comum (saiba mais sobre projeções no Earth Engine ). Podemos definir a resolução (chamada de escala no GEE) sempre que necessário e, claro, temos a opção de não forçar nenhuma reprojeção.

Como você pode ver na descrição dos conjuntos de dados, eles incluem diversos conjuntos de informações armazenados em diversas faixas. Por exemplo, essas faixas estão associadas à coleção LST:

* *LST_Day_1km* : Temperatura da superfície terrestre durante o dia
* *Day_view_time* : hora local de observação do dia
* *LST_Night_1km* : Temperatura da superfície terrestre noturna
etc.

A impressão das primeiras 5 linhas do resultado mostra que agora temos matrizes repletas de dados. Como podemos ver, vários Nonevalores aparecem na coluna LST_Day_1km . O indicador de qualidade associado, QC_Day, indica o valor 2 , o que significa que o LST não é calculado devido aos efeitos das nuvens.

Agora definimos uma função para transformar esse array em um Dataframe pandas, que é muito mais conveniente de manipular.

```python
import pandas as pd

def ee_array_to_df(arr, list_of_bands):
    """Transforma o array retornado pelo `ee.Image.getRegion` no lado do cliente em um `pandas.DataFrame`."""
    df = pd.DataFrame(arr)

    # Reorganize o cabeçalho.
    headers = df.iloc[0]
    df = pd.DataFrame(df.values[1:], columns=headers)

    # Remova as linhas que não contêm dados.
    df = df[['longitude', 'latitude', 'time', *list_of_bands]].dropna()

    # Converta os dados para valores numéricos.
    for band in list_of_bands:
        df[band] = pd.to_numeric(df[band], errors='coerce')

    # Converta o campo de tempo em um formato datetime.
    df['datetime'] = pd.to_datetime(df['time'], unit='ms')

    # Mantenha apenas as colunas de interesse.
    df = df[['time','datetime',  *list_of_bands]]

    return df
```

Aplicamos esta função para obter as duas séries temporais que queremos (e imprimir uma).



```python
lst_df_urban = ee_array_to_df(lst_u_poi,['LST_Day_1km'])

def t_modis_to_celsius(t_modis):
    """Converte as unidades da LST do MODIS para graus Celsius."""
    t_celsius =  0.02*t_modis - 273.15
    return t_celsius

# Aplique a função para obter a temperatura em Celsius.
lst_df_urban['LST_Day_1km'] = lst_df_urban['LST_Day_1km'].apply(t_modis_to_celsius)

# Faça o mesmo para o ponto rural.
lst_df_rural = ee_array_to_df(lst_r_poi,['LST_Day_1km'])
lst_df_rural['LST_Day_1km'] = lst_df_rural['LST_Day_1km'].apply(t_modis_to_celsius)

lst_df_urban.head()
```

Agora que temos nossos dados em bom estado, podemos facilmente criar gráficos e comparar as tendências. Como a região de Lyon, na França, apresenta um clima semicontinental, esperamos que o LST tenha influência sazonal e a tendência sinusoidal descrita por Stallman (1965) , conforme segue:

Captura de tela 2025-10-15 095842.png

Esse código realiza um ajuste de curva senoidal aos dados de temperatura da superfície (LST) para áreas urbanas e rurais, seguindo estes passos:

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy import optimize
%matplotlib inline

# Ajuste de curvas
## Primeiro, extraia os valores de x (tempos) dos dataframes.
x_data_u = np.asanyarray(lst_df_urban['time'].apply(float))  # urban
x_data_r = np.asanyarray(lst_df_rural['time'].apply(float))  # rural

## Em seguida, extraia os valores de y (LST) dos dataframes.
y_data_u = np.asanyarray(lst_df_urban['LST_Day_1km'].apply(float))  # urban
y_data_r = np.asanyarray(lst_df_rural['LST_Day_1km'].apply(float))  # rural

## Depois, defina a função de ajuste com os parâmetros.
def fit_func(t, lst0, delta_lst, tau, phi):
    return lst0 + (delta_lst/2)*np.sin(2*np.pi*t/tau + phi)

## Otimize os parâmetros utilizando um bom palpite inicial `p0`.

lst0 = 20
delta_lst = 40
tau = 365*24*3600*1000   # Milissegundos em um ano
phi = 2*np.pi*4*30.5*3600*1000/tau  # Deslocamento em relação ao momento em que esperamos que LST(t) = LST₀


params_u, params_covariance_u = optimize.curve_fit(
    fit_func, x_data_u, y_data_u, p0=[lst0, delta_lst, tau, phi])
params_r, params_covariance_r = optimize.curve_fit(
    fit_func, x_data_r, y_data_r, p0=[lst0, delta_lst, tau, phi])
```

E agora plote!

```python

# Subplots.
fig, ax = plt.subplots(figsize=(14, 6))

# Adicione gráficos de dispersão.
ax.scatter(lst_df_urban['datetime'], lst_df_urban['LST_Day_1km'],
           c='black', alpha=0.2, label='Urban (data)')
ax.scatter(lst_df_rural['datetime'], lst_df_rural['LST_Day_1km'],
           c='green', alpha=0.35, label='Rural (data)')

# Adicione as curvas de ajuste.

ax.plot(lst_df_urban['datetime'],
        fit_func(x_data_u, params_u[0], params_u[1], params_u[2], params_u[3]),
        label='Urban (fitted)', color='black', lw=2.5)
ax.plot(lst_df_rural['datetime'],
        fit_func(x_data_r, params_r[0], params_r[1], params_r[2], params_r[3]),
        label='Rural (fitted)', color='green', lw=2.5)

# Adicione alguns parâmetros.
ax.set_title('Daytime Land Surface Temperature Near Lyon', fontsize=16)
ax.set_xlabel('Date', fontsize=14)
ax.set_ylabel('Temperature [C]', fontsize=14)
ax.set_ylim(-0, 40)
ax.grid(lw=0.2)
ax.legend(fontsize=14, loc='lower right')

plt.show()
```

### Atividade 1 - Calcular e plotar média anual

**Objetivo**

Conseguir identificar através dos dados e gráficos, os anos mais frios e mais quentes dentro do anos definidos!

```python
# ========================================
# Atividade 1: Calcular e plotar média anual
# ========================================

# Função para calcular média anual de LST
# [APENAS RODE!]
def mean_annual_lst(year):
    start = ee.Date.fromYMD(year, 1, 1)
    end = ee.Date.fromYMD(year, 12, 31)

    collection = ee.ImageCollection(prod) \
        .filterBounds(roi) \
        .filterDate(start, end) \
        .select(data_band)

    # Convertendo de LST MODIS para °C
    annual_mean = collection.mean().multiply(0.02).subtract(273.15)
    return annual_mean.set('year', year)
```

```python
# Calcula imagens médias anuais
# [APENAS RODE!]
annual_images = [mean_annual_lst(y) for y in anos]
```

```python
# Função para calcular média da ROI
# [APENAS RODE!]

def mean_temp(image):
    mean_dict = image.reduceRegion(
        reducer=ee.Reducer.mean(),
        geometry=roi,
        scale=1000,
        maxPixels=1e9
    )
    return ee.Feature(None, {'year': image.get('year'), 'mean_temp': mean_dict.get(data_band)})
```

```python
# Calcula médias da ROI para cada ano
# [APENAS RODE!]

features = [mean_temp(img) for img in annual_images]
fc = ee.FeatureCollection(features)
```

Para cada imagem anual (annual_images), aplica a função mean_temp Cria uma FeatureCollection, que é como uma tabela no Earth Engine: cada linha = ano, cada coluna = média de temperatura.

Agora vamos extrair listas de anos e médias


```python
# Extrair listas
# [APENAS RODE!]

temp_list = fc.aggregate_array('mean_temp').getInfo() # aggregate_array('mean_temp') → cria uma lista Python com as médias anuais.
year_list = fc.aggregate_array('year').getInfo() # aggregate_array('year') → cria uma lista com os anos correspondentes.
                                                  # getInfo() → converte do formato Earth Engine para uma lista Python.
```

Identificando anos mais frio e mais quente

### Atividade 2 – Análise Sazonal (média mensal).
**Objetivo**

Analisar como a temperatura varia ao longo dos meses de um ano, ou seja, capturar a sazonalidade.

```python
# ========================================
# Atividade 2: Análise Sazonal (média mensal)
# ========================================
# Exemplo para 1 ano
# Define o ano que você quer analisar.
# Cria a lista de meses de 1 a 12.
# Inicializa uma lista vazia monthly_means para armazenar a média de cada mês.

year = 2017 # Escolha o ano
months = list(range(1,13))
monthly_means = []
```

Para cada mês:


```python
# [APENAS RODE]
for m in months:
    start = ee.Date.fromYMD(year, m, 1) #  start → primeiro dia do mês
    end = start.advance(1, 'month') # end → primeiro dia do mês seguinte

    # Isso cria um intervalo mensal para filtrar os dados MODIS. Filtra a coleção MODIS LST para a ROI e para o mês específico.
    collection = ee.ImageCollection(prod).filterBounds(roi).filterDate(start,end).select(data_band)

    # Calcula a média das imagens do mês.
    monthly_mean = collection.mean().multiply(0.02).subtract(273.15)

    # Multiplica por 0.02 e subtrai 273.15 para converter de Kelvin para °C.
    # Calcula a média espacial da ROI para o mês usando reduceRegion.
    mean_dict = monthly_mean.reduceRegion(reducer=ee.Reducer.mean(), geometry=roi, scale=1000, maxPixels=1e9)

    # Adiciona o valor à lista monthly_means
    monthly_means.append(mean_dict.get(data_band).getInfo())
```

Plote um gráfico de linha da temperatura média mensal:





```python
# Plota um gráfico de linha da temperatura média mensal:

plt.figure(figsize=(10,6))
plt.plot(months, monthly_means, marker='o', linestyle='-', color='green') # Configure as cores como desejar
plt.title(f'Média Mensal de Temperatura - {year}')
plt.xlabel('Mês') # Eixo X → meses (1 a 12)
plt.ylabel('Temperatura (°C)') # Eixo Y → temperatura média (°C) Visualiza picos e vales de temperatura ao longo do ano.
plt.grid(True)
plt.show()
```

### Atividade 3 -  Tendência Temporal

**Objetivo**

Mostrar se a temperatura média anual está aumentando, diminuindo ou está estável ao longo dos anos, ou seja, calcular uma linha de tendência (regressão linear) sobre os dados anuais.

```python
# ========================================
# Atividade 3: Tendência Temporal
# ========================================

# [APENAS RODE!]
import numpy as np
from numpy.polynomial.polynomial import polyfit

# Ajuste linear
coef = polyfit(year_list, temp_list, 1)  # coef[0]=intercepto, coef[1]=inclinação
trend = [coef[1]*y + coef[0] for y in year_list]

```

```python
# Plote a média anual com o dado year_list

plt.figure(figsize=(10,6))

plt.plot(year_list,
         temp_list,
         marker='o',
         linestyle='-',
         color='orange', # Configure as cores como desejar
         label='Média anual')
plt.plot(year_list,
         trend,
         color='red', # Configure as cores como desejar
         linestyle='--',
         label=f'Tendência ({coef[1]:.3f} °C/ano)')
plt.title('Tendência de Temperatura Anual')
plt.xlabel('Ano')
plt.ylabel('Temperatura (°C)')
plt.grid(True)
plt.legend()
plt.show()
```
